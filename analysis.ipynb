```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Math Avg:", df["math score"].mean())
print("Reading Avg:", df["reading score"].mean())
print("Writing Avg:", df["writing score"].mean())
```

```python
df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
```

```python
df.groupby("gender")[["math score", "reading score", "writing score"]].mean().plot(kind="bar")
plt.title("Average Scores by Gender")
plt.ylabel("Scores")
plt.show()
```

```python
df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
```

```python
df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean().plot(kind="bar")
plt.title("Effect of Test Preparation")
plt.ylabel("Scores")
plt.show()
```
